```
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0,5,0.1)  #starting point, end point, intervals
y = np.sin(x)
plt.plot(x,y) #plot the graph based on y
```

```
import numpy as np
import matplotlib.pyplot as plt
speed = np.array([1,6,4,15,-6,3])
plt.plot(speed)     #plot only based on y, y in this case is speed, x is default
plt.ylabel('Speed') #labels x-axis
plt.xlabel('Moves') #labels y-axis
plt.show() #what does it do?
```

```
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[1,4,9,16],'ro')  #coordinates of points
plt.axis([0,6,0,20]) #axis dimensions
plt.show() #display all figures and return to the ipython prompt.
```
