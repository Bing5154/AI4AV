# Convolutions with special kernels

```
I = np.array([[0,0,0],[0,1,0],[0,0,0]]) #identity matrix as kernel
img_I = signal.convolve2d(img, I, mode='valid') 
plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title("original image")
plt.subplot(1,2,2)
plt.imshow(img_I, cmap='gray')
plt.title("image after convolution")
```

```
I = np.array([[1,0,-1],[2,0,-2],[1,0,-1]]) #kernel, edge detection(vertical)
img_I = signal.convolve2d(img, I, mode='valid') 
plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title("original image")
plt.subplot(1,2,2)
plt.imshow(img_I, cmap='gray')
plt.title("image after convolution")
```

```
I = np.array([[1,2,1],[0,0,0],[-1,-2,-1]]) #kernel, edge detection(horizontal)
img_I = signal.convolve2d(img, I, mode='valid') #valid 2d convolution using identity matrix
plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title("original image")
plt.subplot(1,2,2)
plt.imshow(img_I, cmap='gray')
plt.title("image after convolution")
```

```
I = np.array([[1/16, 1/8, 1/16],[1/8,1/4,1/8],[1/16,1/8,1/16]]) #
img_I = signal.convolve2d(img, I, mode='valid') 
plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title("original image")
plt.subplot(1,2,2)
plt.imshow(img_I, cmap='gray')
plt.title("image after convolution")
```

```
I = np.array([[1/16, 1/8, 1/16],[1/8,1/4,1/8],[1/16,1/8,1/16]]) #kernel: Gaussian blur, cannot be observed by eye
img_I = signal.convolve2d(img, I, mode='same') # The output is the same size as in1 (aka. input)
img_K = img_I - img

plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title("original image")

plt.subplot(1,2,2)
plt.imshow(img_K, cmap='gray')
plt.title("difference for image 3 (the last one): ")

```
